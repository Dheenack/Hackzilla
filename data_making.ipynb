```python
import pandas as pd
```

# datas

```python
csv1=pd.read_csv("data/anemia (1).csv")
# csv2=pd.read_csv("data/Anemia_processed.csv")
csv2=pd.read_csv("data/anemia.csv")
# csv4=pd.read_csv("data/CBC data_for_meandeley_csv (1).csv")
# csv5=pd.read_csv("data/CBC data_for_meandeley_csv.csv")
csv3=pd.read_csv("data/combined_anemia.csv")
csv4=pd.read_csv("data/csvconverted.csv")
csv5=pd.read_csv("data/predicted_labels (1).csv")
```

```python
j=1
for i in [csv1,csv2,csv3,csv4,csv5]:
    print(f"Data {j}");
    j+=1
    display(i.head())
```

```python
csv3=csv3.drop(columns=["Age","PCV","RBC","source"],axis=1)
```

```python
csv4=csv4.drop(columns=["Age","RBC","PCV"],axis=1)
```

```python
j=1
for i in [csv1,csv2,csv3,csv4,csv5]:
    print(f"Data {j}");
    j+=1
    display(i.head())
```

```python
csv3 = csv3.rename(columns={'Hb': 'Hemoglobin', 'Decision_Class': 'Result'})
```

```python
csv4 = csv4.rename(columns={'Hb': 'Hemoglobin', 'Decision_Class': 'Result'})
```

```python
j=1
for i in [csv1,csv2,csv3,csv4,csv5]:
    print(f"Data {j}");
    j+=1
    display(i.head())
```

```python
csv3["Gender"]=csv3["Gender"].map({"f":0,"m":1})
csv3.head()
```

```python
csv4["Gender"]=csv4["Gender"].map({"f":0,"m":1})
csv4.head()
```

```python
j=1
for i in [csv1,csv2,csv3,csv4,csv5]:
    print(f"Data {j}");
    j+=1
    display(i.head())
```

```python
columns=["Gender","Hemoglobin","MCH","MCHC","MCV"]
X_train=pd.concat([csv1[columns],csv2[columns],csv3[columns],csv4[columns],csv5[columns]],ignore_index=True)
display(X_train.head())
print(X_train.shape)
```

```python
X_train.to_csv("Train_data.csv")
```

```python
X_train_target=pd.concat([csv1["Result"],csv2["Result"],csv3["Result"],csv4["Result"],csv5["Result"]],ignore_index=True)
display(X_train_target.head())
```
